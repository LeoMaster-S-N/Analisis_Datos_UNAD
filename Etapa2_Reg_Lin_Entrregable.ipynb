Importar librerías

```python
import matplotlib.pyplot as plt
from sklearn import linear_model
import seaborn as sns
import numpy as np
import pandas as pd
```

Cargar la info del csv

```python
rl = pd.read_csv ("C:\Jupyter Dats\Datasets\RL Ejercicio\data.csv", sep=",")
```

```python
plt.plot(rl)
plt.title("Gráfico de Líneas")
plt.xlabel("Metros")
plt.ylabel("Precios")
```

Crear gráfica de dispersión

```python
rl.plot.scatter(x="metro", y="precio")
plt.show()
```

Crear variable de Regresión Líneal

```python
reglin = linear_model.LinearRegression()
```

Agregar datos al array

```python
metros = rl["metro"].values.reshape((-1,1))
```

Crear variable modelo para los datos

```python
modelo = reglin.fit(metros, rl["precio"])
```

Imprimir Intersección y Pendiente de los datos

```python
print("Intersección (b)", modelo.intercept_)
print("Pendiente (m)", modelo.coef_)
```

Crear variables contenedoras de arrays de entrada y predicciones para graficar linealmente

```python
entrada = [[5],[15],[20],[25]]
predicciones = modelo.predict(entrada)
```

Imprimir la variable predicciones

```python
print(predicciones)
```

Mostrar los resultados en la gráfica de regresión líneal

```python
rl.plot.scatter(x="metro", y="precio", label = "Información")
plt.scatter(entrada, predicciones, color = 'orange')
plt.plot(entrada, predicciones, color = 'green', label = "Línea de Regresión")
plt.xlabel("Metros")
plt.ylabel("Precios")
plt.title("Mí Regresión Líneal Etapa 2")
plt.legend()
plt.show()
```
